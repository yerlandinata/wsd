# Rule-Based MWE Detection

```python
import requests
import json
import time
```

```python
test = dataset['kata'].unique()
test
```

```python
mwe_dict = dict()
for word in dataset['kata'].unique():
    api_url = 'https://kateglo.com/api.php?'
    api_param = {'format':'json', 'phrase':word}
    data = requests.get(api_url, api_param).json()
    related_phrases = data['kateglo']['relation']['c']
    mwe_dict[word] = []
    for phrase in related_phrases.values():
        if type(phrase) == dict:
            mwe_dict[word].append(phrase['related_phrase'])
    time.sleep(1) #Dealing with Kateglo API rate limit
```

```python
mwe_dict
```

```python
def checkMWE(sentence, word):
    for phrase in mwe_dict[word]:
        if phrase in sentence:
            return phrase
```

```python
dataset['has_mwe'] = list(map(
    lambda word, sentence: True if checkMWE(sentence, word) else False, dataset.kata, dataset.kalimat))
dataset['mwe'] = list(map(lambda word, sentence: checkMWE(sentence, word), dataset.kata, dataset.kalimat))
```

```python
truePos = dataset.loc[(dataset.has_mwe == True) & (dataset.sense.str.contains('0x'))]
falsePos = dataset.loc[(dataset.has_mwe == True) & (~dataset.sense.str.contains('0x'))]
trueNeg = dataset.loc[(dataset.has_mwe == False) & (~dataset.sense.str.contains('0x'))]
falseNeg = dataset.loc[(dataset.has_mwe == False) & (dataset.sense.str.contains('0x'))]
precision = len(truePos)/(len(truePos)+len(falseNeg))
recall = len(truePos)/(len(truePos)+len(falsePos))
fscore = 2*precision*recall/(recall + precision)

print('TP =', len(truePos))
print('FP =', len(falsePos))
print('TN =', len(trueNeg))
print('FN =', len(falseNeg))
print('Precision =', precision)
print('Recall =', recall)
print('F-Score =', fscore)
```

Our rule-based MWE detector Has a really low recall even though we use dictionary as a reference for our corpus. The reason is because there are a lot of inconsistency in this data.

## Execute cleaning

```python
import time
from preprocessor import (
    normalize_money, normalize_number, remove_punctuation,
    create_stemmer, create_stop_words_remover, stemmer,
    clean_word, normalize_weekday, normalize_month, remove_normalized,
    normalize_personal_pronoun, normalize_demonstrative_pronoun, normalize_coordinating_conjunction,
    normalize_determiner, normalize_preposition, create_obvious_verb_normalizer
)
```

```python
# exceptions, do not stem, because pysastrawi is stupid
exception_words = {'senilai', 'bernilai', 'menilainya', 'dinilainya', *set(map(stemmer.stem, annotated_words))}
```

```python
begin = time.perf_counter()
dataset['kalimat'] = pd.Series(
    map(
        pipe(
            remove_punctuation,
            normalize_money,
            normalize_number,
            normalize_weekday,
            normalize_month,
            create_stemmer(annotated_words, exception_words),
        ),
        dataset.kalimat
    )
)
print('elapsed time:', time.perf_counter() - begin)
```

```python
begin = time.perf_counter()
dataset['clean'] = pd.Series(
    map(
        pipe(
            create_obvious_verb_normalizer(annotated_words, exception_words),
            normalize_personal_pronoun,
            normalize_demonstrative_pronoun,
            normalize_coordinating_conjunction,
            normalize_determiner,
            normalize_preposition,
            remove_normalized,
            create_stop_words_remover(annotated_words, exception_words),
        ),
        dataset.kalimat
    )
)
print('elapsed time:', time.perf_counter() - begin)
```

# Write down the position of target word

```python
targetpos = [-1 for i in range(len(dataset))]

for i in range(len(dataset)):
    target = dataset.iloc[i].kata
    tokens = dataset.iloc[i].clean.split()
    for j in range(len(tokens)):
        if tokens[j].find(target) >= 0 or stemmer.stem(tokens[j]) == stemmer.stem(target):
            targetpos[i] = j
            break
    if targetpos[i] == -1:
        print(target)
        print(tokens)
        print(dataset.iloc[i].kalimat)
        print(i)
        raise Exception

dataset['targetpos_clean'] = pd.Series(targetpos)
```

```python
targetpos_ori = [-1 for i in range(len(dataset))]

for i in range(len(dataset)):
    target = dataset.iloc[i].kata
    tokens = dataset.iloc[i].kalimat.split()
    for j in range(len(tokens)):
        token = clean_word(tokens[j])
        if token.find(target) >= 0 or stemmer.stem(token) == stemmer.stem(target):
            targetpos_ori[i] = j
            break
    if targetpos_ori[i] == -1:
        print(target)
        print(tokens)
        print(dataset.iloc[i].kalimat)
        print(i)
        raise Exception

dataset['targetpos_ori'] = pd.Series(targetpos_ori)
```

```python
targetpos_pos_tag = [-1 for i in range(len(dataset))]
tags_only = ['' for i in range(len(dataset))]

for i in range(len(dataset)):
    target = dataset.iloc[i].kata
    tokens = dataset.iloc[i].pos_tags
    for j in range(len(tokens)):
        token = clean_word(tokens[j][0])
        if token.find(target) >= 0 or stemmer.stem(token) == stemmer.stem(target):
            targetpos_pos_tag[i] = j
            break
    if targetpos_pos_tag[i] == -1:
        print(target)
        print(tokens)
        print(dataset.iloc[i].kalimat)
        print(i)
        raise Exception
    tags_only[i] = ' '.join(list(map(lambda t: t[1], tokens)))

dataset['targetpos_pos_tag'] = pd.Series(targetpos_pos_tag)
dataset['pos_tags'] = pd.Series(tags_only)
```

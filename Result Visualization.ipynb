```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

# Load Training Results

```python
res = pd.read_csv('iacobacci.csv').sort_values(by=['5-fold_validation_f1_macro'])
res.head()
```

# Visualize

## Iacobacci

```python
sorted([2,3,1], key=lambda i: -i)
```

```python
# data to plot
n_groups = 4
train = res['train_f1_macro']
validation = res['5-fold_validation_f1_macro']
dummy = res['dummy_f1_macro']

# create plot
fig, ax = plt.subplots(figsize=(15,50))
index = np.arange(len(res))
bar_width = 0.35
opacity = 0.8

rects1 = plt.barh(
    2*index + 2*bar_width, train, bar_width,
    alpha=opacity-.3,
    color='g',
    label='Training Macro F1'
)

rects2 = plt.barh(
    2*index + bar_width, validation, bar_width,
    alpha=opacity,
    color='r',
    label='5-Fold Validation Macro F1'
)

rects3 = plt.barh(
    2*index, dummy, bar_width,
    alpha=1.0,
    color='y',
    label='Dummy Macro F1'
)

plt.xlabel('Scores')
plt.ylabel('Ambiguous words')
plt.title('Scores by ambiguous words')
plt.yticks(2*index + bar_width, res.word)
plt.legend()

plt.tight_layout()
plt.show()
```

# Compare and Put this on the paper

```python
res_a = pd.read_csv('iacobacci_ims_full_embedding400.csv')
res_b = pd.read_csv('yerlandinata_word_expert.csv')

# data to plot
n_groups = 4
a = res_a['5-fold_validation_f1_macro']
b = res_b['5-fold_validation_f1_macro']

# create plot
fig, ax = plt.subplots(figsize=(15,30))
index = np.arange(len(a))
bar_width = 0.35
opacity = 0.9

rects1 = plt.barh(
    2*index + 2*bar_width, a, bar_width,
    alpha=opacity,
    color='y',
    label='Iacobacci'
)

rects2 = plt.barh(
    2*index + bar_width, b, bar_width,
    alpha=opacity,
    color='r',
    label='Word Expert'
)

plt.xlabel('Scores')
plt.ylabel('Ambiguous words')
plt.title('Scores by ambiguous words')
plt.yticks(2*index + bar_width, res_a.word)
plt.legend()

plt.tight_layout()
plt.show()
```
